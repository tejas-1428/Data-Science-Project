```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
Data = pd.read_csv("../airline/wisc_bc_data.csv")
```

```python
X = Data[['radius_mean','radius_se','concave points_mean']]
y = Data['diagnosis']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=10,shuffle=True,train_size=0.2)
```

```python
# there are other sampling method also like KFold
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=5)
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix,accuracy_score,balanced_accuracy_score,f1_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
model.score(X_test,y_test)
```

```python
balanced_accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test,y_pred)
```
